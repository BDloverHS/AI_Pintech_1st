```python
from faker import Faker
```

```python
faker = Faker() # 기본값은 영어 데이터
faker.name()
```

```python
faker = Faker("ko_KR") # 한국어로 설정
faker.name()
```

```python
members = [(faker.email(), faker.name(), faker.address())
           for _ in range(0, 1000)]
```

```python
for member in members:
    print(member)
```

```python
members = [faker.first_name() for _ in range(0, 1000)]

for member in members:
        print(member)
```
