```python
from tensorflow import keras # keras API
```

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
print(train_input.shape, test_input.shape)
# 60000개의 훈련 데이터셋(28*28 사이즈)과 10000개의 테스트 데이터셋(28*28 사이즈)
```

```python
import matplotlib.pyplot as plt

# subplots -> 그래프 영역을 여러 개 나눠서 출력
fig, axs = plt.subplots(1, 10, figsize=(10,10)) # 1행 10열, figsize : 이미지의 크기 조절
for i in range(10):
    axs[i].imshow(train_input[i], cmap='gray_r') # 이미지를 보여주는 메서드  / cmap='색상_r' : 반전. 물건이 중요하므로 물건이 잘 보이도록.
    axs[i].axis('off') # 좌표 출력 배제

plt.show()
```

```python
print(train_target[:10])
```

```python
# 확률적 경사 하강법
# 0 ~ 255
# print(train_input[:5])

train_scaled = train_input / 255 # 0 혹은 1이 나오도록
train_scaled = train_scaled.reshape(-1, 28 * 28) # reshape : 데이터 가공 / 특징을 78행 78열로 바꿈
```

```python
print(train_scaled.shape)
```

```python
test_scaled = test_input / 255 # 0~1
test_scaled = test_scaled.reshape(-1, 28 * 28)
```

```python
print(test_scaled.shape)
```

```python
from sklearn.linear_model import SGDClassifier

sc = SGDClassifier(loss = 'log_loss', max_iter=5)
```

```python
sc.fit(train_scaled, train_target) # 훈련
```

```python
sc.score(test_scaled, test_target)
```

```python
sc.predict(test_scaled[:5])
```

```python
test_input2 = test_scaled.reshape(-1, 28 * 28)
```

```python
fig, axs = plt.subplots(1, 5, figsize=(10,10))
for i in range(5):
    axs[i].imshow(test_input2[i], cmap='gray_r')
    axs[i].axis('off')

plt.show()
```

```python
# 텐서플로 + 케라스 - 심층 신경망(딥러닝)
input = keras.layers.Input(shape=(784,)) # 입력층

dense = keras.layers.Dense(10, activation="softmax") # 출력층, 활성화 함수 - softmax : 마지막층은 분류 확률을 구하는 것이기 때문.
```

```python
# model = keras.Sequential([input, dense])
model = keras.Sequential()
model.add(input)
model.add(dense)
```
